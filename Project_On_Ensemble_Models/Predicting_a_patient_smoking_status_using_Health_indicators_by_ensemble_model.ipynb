```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier,StackingClassifier
import xgboost as xgb
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,precision_recall_curve,roc_auc_score,roc_curve,explained_variance_score
```

```python
x=df_train.drop(['smoking'],axis=1)
y=df_train['smoking']
```

```python
pca=PCA(n_components=15,random_state=42)
x_pca=pca.fit_transform(x)
print(f"Explained Variance Ratio:{pca.explained_variance_ratio_}")
print(f"Cummulative Variance Explained:{sum(pca.explained_variance_ratio_):.2f}")
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
rf_clf=RandomForestClassifier(n_estimators=100,random_state=42)
rf_clf.fit(x_train,y_train)
rf_params={'n_estimators':[100,200],'criterion':['entropy','log_loss'],'max_depth':[1,2,3],'max_features':['sqrt','log2']}
rf_grid=GridSearchCV(rf_clf,param_grid=rf_params,cv=6,return_train_score=True,scoring='accuracy')
rf_grid.fit(x_train,y_train)
rf_pred=rf_grid.predict(x_test)
rf_best_score=rf_grid.best_score_
rf_best_params=rf_grid.best_params_
print(f"Classification Report:\n{classification_report(y_test,rf_pred)}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test,rf_pred)}")
print(f"Accuracy Score:{accuracy_score(y_test,rf_pred)}")
print(f"Best Score:",rf_best_score)
print(f"Best Params:",rf_best_params)
```

```python
rf_pred_proba=rf_grid.predict_proba(x_test)[:,1]
precision,recall,thresholds=precision_recall_curve(y_test,rf_pred_proba)
fpr,tpr,thresholds=roc_curve(y_test,rf_pred_proba)
auc=roc_auc_score(y_test,rf_pred_proba)
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(12,6))
axes[0].plot(recall,precision,label='Random Forest Classifier')
axes[0].set_title('Precision-Recall Curve')
axes[0].set_xlabel('Recall')
axes[0].set_ylabel('precision')
axes[0].legend()
axes[1].plot(fpr,tpr,label=f'ROC Curve(AUC:{auc:.2f})',color='blue')
axes[1].set_title('Receiver Operating Characteristic Curve')
axes[1].set_xlabel('False Positive Rate')
axes[1].set_ylabel('True Positive Rate')
axes[1].legend()
plt.tight_layout()
plt.show()
```

```python
xgb_clf=xgb.XGBClassifier(random_state=42)
xgb_clf.fit(x_train,y_train)
xgb_params={'max_depth':[5,6,7],'learning_rate':[0.1,0.3],'n_estimators':[100,200],'reg_alpha':[0,0.01],
'reg_lambda':[1,0.8]}
xgb_grid=GridSearchCV(xgb_clf,param_grid=xgb_params,cv=6,return_train_score=True,scoring='accuracy')
xgb_grid.fit(x_train,y_train)
xgb_pred=xgb_grid.predict(x_test)
xgb_best_score=xgb_grid.best_score_
xgb_best_params=xgb_grid.best_params_
print(f"Classification Report:\n{classification_report(y_test,xgb_pred)}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test,xgb_pred)}")
print(f"Accuracy Score:{accuracy_score(y_test,xgb_pred)}")
print(f"Best Score:",xgb_best_score)
print(f"Best Params:",xgb_best_params)
```

```python
xgb_pred_proba=xgb_grid.predict_proba(x_test)[:,1]
precision,recall,thresholds=precision_recall_curve(y_test,xgb_pred_proba)
fpr,tpr,thresholds=roc_curve(y_test,xgb_pred_proba)
auc=roc_auc_score(y_test,xgb_pred_proba)
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(12,6))
axes[0].plot(recall,precision,label='XG Boosting Classifier')
axes[0].set_title('Precision-Recall Curve')
axes[0].set_xlabel('Recall')
axes[0].set_ylabel('precision')
axes[0].legend()
axes[1].plot(fpr,tpr,label=f'ROC Curve(AUC:{auc:.2f})',color='blue')
axes[1].set_title('Receiver Operating Characteristic Curve')
axes[1].set_xlabel('False Positive Rate')
axes[1].set_ylabel('True Positive Rate')
axes[1].legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
```

```python
lr_clf=LogisticRegression(random_state=42)
lr_params={'penalty':['l1','l2','elasticnet'],'max_iter':[100,150]}
lr_grid=GridSearchCV(lr_clf,cv=6,scoring='accuracy',param_grid=lr_params)

nav_clf=GaussianNB()
nav_params={'priors':[None]}
nav_grid=GridSearchCV(nav_clf,cv=6,param_grid=nav_params,return_train_score=True,scoring='accuracy')

dt_clf=DecisionTreeClassifier(criterion='gini',max_depth=4)
dt_params={'criterion':['gini','entropy','log_loss'],'max_depth':[1,5,7],'splitter':['best','random'],'max_features':['auto','log2','sqrt'],
'max_leaf_nodes':[1,2]}
dt_grid=GridSearchCV(dt_clf,param_grid=dt_params,cv=6,scoring='accuracy')


ensemble_clf=VotingClassifier(estimators=[('lr',lr_grid),('nav',nav_grid),('dt',dt_grid)],weights=[1,1,2])
ensemble_params={'voting':['hard','soft']}
ensemble_grid=GridSearchCV(ensemble_clf,cv=6,param_grid=ensemble_params,scoring='accuracy')
ensemble_grid.fit(x_train,y_train)
ensemble_pred=ensemble_grid.predict(x_test)
print(f"Classification Report:\n{classification_report(y_test,ensemble_pred)}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test,ensemble_pred)}")
print(f"Accuracy Score:{accuracy_score(y_test,ensemble_pred)}")
```

```python
ensemble_pred_proba=ensemble_grid.predict_proba(x_test)[:,1]
precision,recall,thresholds=precision_recall_curve(y_test,ensemble_pred_proba)
fpr,tpr,thresholds=roc_curve(y_test,ensemble_pred_proba)
auc=roc_auc_score(y_test,ensemble_pred_proba)
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(12,6))
axes[0].plot(recall,precision,label='Voting Classifier')
axes[0].set_title('Precision-Recall Curve')
axes[0].set_xlabel('Recall')
axes[0].set_ylabel('precision')
axes[0].legend()
axes[1].plot(fpr,tpr,label=f'ROC Curve(AUC:{auc:.2f})',color='blue')
axes[1].set_title('Receiver Operating Characteristic Curve')
axes[1].set_xlabel('False Positive Rate')
axes[1].set_ylabel('True Positive Rate')
axes[1].legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import StackingClassifier
```

```python
lg_clf=LogisticRegression(random_state=42)
lg_params={'penalty':['l1','l2','elasticnet'],'multi_class':['auto','ovr']}
lg_grid=GridSearchCV(lg_clf,param_grid=lg_params,cv=6,scoring='accuracy')

nav_clf=GaussianNB()
nav_params={'priors':[None]}
nav_grid=GridSearchCV(nav_clf,cv=6,param_grid=nav_params,return_train_score=True,scoring='accuracy')


dt_clf=DecisionTreeClassifier(criterion='gini',max_depth=4)
dt_params={'criterion':['gini','entropy','log_loss'],'max_depth':[1,5,7],'splitter':['best','random'],'max_features':['auto','log2','sqrt'],
'max_leaf_nodes':[1,2]}
dt_grid=GridSearchCV(dt_clf,param_grid=dt_params,cv=6,scoring='accuracy')

base_model=[('lg',lg_grid),('nav',nav_grid)]

meta_model=dt_grid

stacking_clf=StackingClassifier(estimators=base_model,final_estimator=meta_model)

stacking_clf.fit(x_train,y_train)
stk_pred=stacking_clf.predict(x_test)
print(f"Classification Report:\n{classification_report(y_test,stk_pred)}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test,stk_pred)}")
print(f"Accuracy Score:{accuracy_score(y_test,stk_pred)}")
```

```python
stk_pred_proba=stacking_clf.predict_proba(x_test)[:,1]
precision,recall,thresholds=precision_recall_curve(y_test,stk_pred_proba)
fpr,tpr,thresholds=roc_curve(y_test,stk_pred_proba)
auc=roc_auc_score(y_test,stk_pred_proba)
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(12,6))
axes[0].plot(recall,precision,label='stacking Classifier')
axes[0].set_title('Precision-Recall Curve')
axes[0].set_xlabel('Recall')
axes[0].set_ylabel('precision')
axes[0].legend()
axes[1].plot(fpr,tpr,label=f'ROC Curve(AUC:{auc:.2f})',color='blue')
axes[1].set_title('Receiver Operating Characteristic Curve')
axes[1].set_xlabel('False Positive Rate')
axes[1].set_ylabel('True Positive Rate')
axes[1].legend()
plt.tight_layout()
plt.show()
```
