```python
emp=pd.get_dummies(loan_dataset['EmploymentStatus'],prefix='EmploymentStatus',drop_first=True,dtype='int64')
loan_dataset=pd.concat([loan_dataset,emp],axis=1)
edu=pd.get_dummies(loan_dataset['EducationLevel'],prefix='EducationLevel',drop_first=True,dtype='int64')
loan_dataset=pd.concat([loan_dataset,edu],axis=1)
mar=pd.get_dummies(loan_dataset['MaritalStatus'],prefix='MartialStatus',drop_first=True,dtype='int64')
loan_dataset=pd.concat([loan_dataset,mar],axis=1)
Home=pd.get_dummies(loan_dataset['HomeOwnershipStatus'],prefix='HomeOwnershipStatus',drop_first=True,dtype='int64')
loan_dataset=pd.concat([loan_dataset,Home],axis=1)
loan=pd.get_dummies(loan_dataset['LoanPurpose'],prefix='LoanPurpose',drop_first=True,dtype='int64')
loan_dataset=pd.concat([loan_dataset,loan],axis=1)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
numeric_col=loan_dataset.select_dtypes(include=['float64','int64'])
X=sm.add_constant(numeric_col)
vif_data=pd.DataFrame()
vif_data["feature"]=X.columns
vif_data['VIF']=[variance_inflation_factor(X.values,i)for i in range(X.shape[1])]
print(vif_data.head(23))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
loan_dataset.head()
```

```python
x=loan_dataset.drop(['NumberOfDependents','NumberOfOpenCreditLines','NumberOfCreditInquiries','SavingsAccountBalance','CheckingAccountBalance','UtilityBillsPaymentHistory','JobTenure','RiskScore','MartialStatus_Married','MartialStatus_Widowed','HomeOwnershipStatus_Own','HomeOwnershipStatus_Rent','LoanPurpose_Home'],axis=1)
y=loan_dataset['RiskScore']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Linear Regression
lrmodel=LinearRegression()
lrmodel.fit(x_train,y_train)
```

```python
y_pred=lrmodel.predict(x_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
```

```python
print("Mean Squared Error",mse)
print("Mean Absolute Error",mae)
print("R-Squared Error",r2)
```

```python
loan_dataset_error=pd.DataFrame()
loan_dataset_error['y_test']=y_test
loan_dataset_error['y_pred']=y_pred
```
